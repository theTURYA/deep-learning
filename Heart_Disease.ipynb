## Work Flow:

It's a Binary classification problem.

Heart Data --> Data Pre-processing --> Train- Test Split --> Logic Regression Model

New data --> Trained Logistic Regression Model --> Healthy (or) Heart Defect


Importing the required Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# Loading the csv data to a pandas DataFrame
heart_database = pd.read_csv("/content/heart_disease_data.csv")
heart_database
```

```python
heart_database.head()
```

```python
# Print last 5 rows of the dataset
heart_database.tail()
```

```python
heart_database.shape, heart_database.size
```

```python
heart_database.info()
```

```python
# Checking for missing values
heart_database.isnull().sum()
```

```python
# Statistical measures about the data
heart_database.describe()
```

```python
# Checking the distribution of Target Variable
heart_database['target'].value_counts()
```

Splitting the features and targets

```python
X = heart_database.drop(columns = 'target', axis = 1)
y = heart_database['target']
X, y
```

Splitting the data into training data and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train, X_test
```

```python
y_train, y_test
```

Model Training
* Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the LogisticRegression model with Training data
model.fit(X_train, y_train)
```

Model Evaluation
 * Accuracy Score

```python
# Accuracy on training data
X_train_preds = model.predict(X_train)
training_data_accur = accuracy_score(X_train_preds, y_train)
training_data_accur
```

```python
print(f"Accuracy on Training data: {training_data_accur*100}%")
```

```python
# Accuracy on test data
X_test_preds = model.predict(X_test)
test_data_accur = accuracy_score(X_test_preds, y_test)
test_data_accur
```

```python
print(f"Accuracy on Training data: {test_data_accur*100}%")
```

Building a predictive system

```python
input_data = (68,1,0,144,193,1,1,141,0,3.4,1,2,33)

# Change the input_data to numpy array
numpy_arr = np.asarray(input_data)

# Reshape the numpy array as we are predicting for only one instance
input_data_reshape = numpy_arr.reshape(1,-1)

prediction = model.predict(input_data_reshape)
print(prediction)

if(prediction[0] == 0):
  print("The Person does not have a Heart Disease")
else:
   print("The Person have a Heart Disease")
```
