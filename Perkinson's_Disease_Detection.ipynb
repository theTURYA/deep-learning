## Work Flow:

Parkinson's Data --> Data Pre Processing --> Train Test Split --> Support Vector Machine Classifier

* After training the data:

New Data --> Trained Support Vector Machine Classifier --> Parkinson's (or) Healthy [Prediction]

Importing the required dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn import svm
```

Attribute Information:
* Matrix column entries (attributes):
* name - ASCII subject name and recording number
* MDVP:Fo(Hz) - Average vocal fundamental frequency
* MDVP:Fhi(Hz) - Maximum vocal fundamental frequency
* MDVP:Flo(Hz) - Minimum vocal fundamental frequency
* MDVP:Jitter(%),MDVP:Jitter(Abs),MDVP:RAP,MDVP:PPQ,Jitter:DDP - Several
measures of variation in fundamental frequency
* MDVP:Shimmer,MDVP:Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,MDVP:APQ,Shimmer:DDA - Several measures of variation in amplitude
* NHR,HNR - Two measures of ratio of noise to tonal components in the voice
* **status - Health status of the subject (one) - Parkinson's, (zero) - healthy**
* RPDE,D2 - Two nonlinear dynamical complexity measures
* DFA - Signal fractal scaling exponent
* spread1,spread2,PPE - Three nonlinear measures of fundamental frequency variation

Data Pre-Processing

* Separating the features and targets

```python
X = parkinsons_data.drop(columns=['name', 'status'], axis = 1)
y = parkinsons_data['status']
X, y
```

Splitting the data to Training data and Test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

Model Training
* Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
```

```python
# Training the SVM model with training data
model.fit(X_train, y_train)
```

Model Evaluation

* Accuracy score

```python
# Accuracy score on training data
X_train_preds = model.predict(X_train)
training_data_accur = accuracy_score(y_train, X_train_preds)
```

```python
print(f"Accuracy score of the training data is : {training_data_accur}")
```

```python
# Accuracy score on test data
X_test_preds = model.predict(X_test)
test_data_accur = accuracy_score(y_test, X_test_preds)
```

```python
print(f"Accuracy score of the test data is : {test_data_accur}")
```

Building a Predictive System

```python
input_data = (119.99200,157.30200,74.99700,0.00784,0.00007,0.00370,0.00554,0.01109,0.04374,0.42600,0.02182,0.03130,0.02971,0.06545,0.02211,21.03300,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654
)

# Changing the data to a numpy array
input_data_arr = np.asarray(input_data)

# Reshape the numpy array
input_data_reshaped = input_data_arr.reshape(1,-1)


# Standardize the data
std_data = scaler.transform(input_data_reshaped)
std_data

# Prediction of output
prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 0):
   print("The Person does not have Perkinson's Disease")
else:
   print("The Person have Perkinson's Disease")

```
