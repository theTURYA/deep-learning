## Work Flow:
Diabetes Data --> Data Pre-Processing -->  Train-Test Data --> Support Vector Machine Classifier

* Prediction:

New Data --> Trained Support Vector Machine Classifier --> Diabetic (or) Non-Diabetic

Importing the required Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis
PIMA Diabetes Dataset

```python
# Loading the dataset to pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
diabetes_dataset
```

```python
# Printing the first 5 rows of the dataset
diabetes_dataset.head()
```

```python
# Number of Rows and Columns in this dataset
diabetes_dataset.shape
```

```python
# Getting the sytatistical measures of the data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0 --> Non-Dianetic

1 --> Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Separating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis = 1)
Y = diabetes_dataset['Outcome']
X, Y
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
stand_data = scaler.transform(X)
stand_data
```

```python
X = stand_data
y = diabetes_dataset['Outcome']
X, y
```

Train Test Split


```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2,
                                                    stratify = y, random_state = 2)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
y.shape, y_train.shape, y_test.shape
```

Training the model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
# Training the Support Vector Machine (SVM) Classifier
classifier.fit(X_train, y_train)
```

Model Evaluatation

* Accuracy Score

```python
# Accuracy score on the training data
X_train_preds = classifier.predict(X_train)
training_data_accur = accuracy_score(X_train_preds, y_train)
print(f"Accuracy score of the training data is : {training_data_accur}")
```

```python
#Accuracy score on the test data
X_test_preds = classifier.predict(X_test)
test_data_accur = accuracy_score(X_test_preds, y_test)
print(f"Accuracy score of the training data is : {test_data_accur}")
```

Making a Predictive System

```python
input_data = (6, 148, 72,	35,	0, 33.6,	0.627,	50)


# Changing the input_data to numpy array
data_to_array = np.array(input_data)

# Reshape the array as we are predicting for one instance
input_data_reshaped = data_to_array.reshape(1, -1)

# Standardize the input data
std_data = scaler.transform(input_data_reshaped)
std_data

```

```python
prediction = classifier.predict(std_data)
prediction

if (prediction[0] == 0):
  print("The person is not Diabetic")
else:
  print("The person is Diabetic")
```
